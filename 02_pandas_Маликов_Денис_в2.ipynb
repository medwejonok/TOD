```python
a = array.copy()
summ = []
x = []
for ticker in a['ticker']:
    if ticker in x:
        continue
    b = a[a['ticker'] == ticker].head()
    summ.append([ticker, sum(b['volume'])])
    x.append(ticker)
summ
```
